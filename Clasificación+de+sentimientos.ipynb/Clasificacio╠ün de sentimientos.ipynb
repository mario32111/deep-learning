# Caso PrГЎctico: ClasificaciГіn de sentimientos

## Conjunto de datos

Este es un conjunto de datos estГЎ formado por 25.000 crГ­ticas de pelГ­culas del IMDB (Internet Movie Database), etiquetadas por sentimiento (positivo/negativo). Las reseГ±as han sido preprocesadas, y cada reseГ±a estГЎ codificada como una lista de Г­ndices de palabras (nГєmeros enteros). Para mayor comodidad, las palabras se indexan por frecuencia general en el conjunto de datos, de modo que, por ejemplo, el nГєmero entero "3" codifica la tercera palabra mГЎs frecuente en los datos. Esto permite realizar operaciones de filtrado rГЎpido como considerar sГіlo las 10.000 palabras mГЎs comunes.

## 1. Cargando el conjunto de datos

```python
from tensorflow.keras.datasets import imdb
```

```python
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)
```

El argumento _num_words_ indica que Гєnicamente tendremos en cuenta las 10.000 palabras mГЎs frecuentes del conjunto de datos. Por lo tanto, el resto de palabras que sean poco frecuentes, serГЎn descartadas y no formarГЎn parte de las reseГ±as con las que entrenaremos y probaremos el algoritmo.

Tanto _train_data_ como _test_data_ se corresponden con una lista de reseГ±as. Cada una de las reseГ±as se corresponde con una lista de nГєmeros enteros que representan el Г­ndice de una palabra. Para conocer que Г­ndice tiene cada palabra, podemos llamar al mГ©todo _get_word_index()_

```python
# ГЌndices de las palabras
word_index = imdb.get_word_index()
word_index
```

```python
# Primera reseГ±a del conjunto de datos de entrenamiento
print(train_data[0])
```

Es importante tener en cuenta que a la hora de visualizar las reseГ±as debemos aplicar un offset de -3 debido a que el nГєmero 0, 1 y 2 estГЎn reservados para "padding", "start of sequence" y "unknown" respectivamente.

```python
# Decodificando la reseГ±a utilizando los Г­ndices de las palabras
reversed_word_index = {v: k for k, v in word_index.items()}
' '.join([reversed_word_index.get(i - 3, "({})".format(i)) for i in train_data[0]])
```

Las variables _train_labels_ y _test_labels_ se corresponden con una lista de valores que puede ser 1, si la reseГ±a es positiva, o 0, si la reseГ±a es negativa

```python
# La reseГ±a anterior se corresponde con una reseГ±a positiva
print(train_labels[0])
```

## 2. Preprocesamiento del conjunto de datos

A pesar de que ya tenemos codificadas las reseГ±as para que esten representadas por valores numГ©ricos en lugar de por cadenas de texto, existen algunos factores que no permiten que sean conjuntos de datos apropiados para proporcionarle a una red neuronal artifical.

Una de las limitaciones fundamentales es que cada reseГ±a tiene un nГєmero diferente de palabras.

```python
import numpy as np

# Vectorizamos las reseГ±as convirtiendolas en vectores de 10.000 valores 1s y 0s
def vectorize(seqs, dim=10000):
    results = np.zeros((len(seqs), dim))
    for i, seq in enumerate(seqs):
        results[i, seq] = 1.
    return results
```

```python
# Para que lo anterior funcione se basa en la siguiente propiedad
a = np.zeros(15)
print("Array original:\t\t", a)
a[[0, 3, 5, 7, 9]] = 1.
print("Array modificado:\t", a)
```

Vectorizamos tanto el conjunto de datos de entrenamiento, como el conjunto de datos de pruebas.

```python
# Vectorizamos el conjunto de datos de entrenamiento y el de pruebas
X_train = vectorize(train_data, 10000)
X_test = vectorize(test_data, 10000)
```

```python
# Cada una de las reseГ±as se corresponde con un vector formado por 1s y 0s
print("Valores originales:\t", train_data[0][:15])
print("Valores vectorizados:\t", X_train[0][:15])
```

## 3. DivisiГіn del conjunto de datos

```python
from sklearn.model_selection import train_test_split

X_test, X_val, Y_test, Y_val = train_test_split(X_test, test_labels, test_size=0.5)
```

```python
Y_train = train_labels
```

```python
print("Longitud subconjunto entrenamiento: ", len(X_train))
print("Longitud subconjunto validaciГіn: ", len(X_val))
print("Longitud subconjunto pruebas: ", len(X_test))
```

## 4. ConstrucciГіn de la red neuronal artificial

```python
from tensorflow.keras import models
from tensorflow.keras import layers

model = models.Sequential()
model.add(layers.Dense(16, activation='tanh', input_shape=(10000,)))
model.add(layers.Dense(16, activation='tanh'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
from tensorflow.keras import optimizers

# Aqui se usa un learning_rate personalizado, no el de 0.1, se debe usar 0.01
# debido a que si se usa el predeterminado se logra apreciar mucho ruido en el error y en el  acuracy, esto debido a qye el valor
# con el que esta actualizando los parametros es muy grande
model.compile(
    optimizer=optimizers.SGD(learning_rate=0.001), 
    loss='binary_crossentropy',
    metrics=['accuracy', 'Precision']
)
```

```python
history = model.fit(
    X_train,
    Y_train,
    epochs=40,
    validation_data=(X_val, Y_val))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history)[['loss', 'val_loss']].plot(figsize=(10, 6))
plt.grid(True)
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history)[['accuracy', 'val_accuracy']].plot(figsize=(10, 6))
plt.grid(True)
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.show()
```

### 4.1 EvaluaciГіn con el subconjunto de pruebas

```python
test_loss, test_acc, test_prec = model.evaluate(X_test, Y_test)
```

```python
print('test_acc:', test_acc)
print('test_prec:', test_prec)
```

## 5. Otros preprocesamientos del conjunto de datos

### Cargando el conjunto de datos

Hay muchas formas de preprocesar la informacion me diante codificacion, anteriormente se uso la vectorizacion, aqui se va a resolver el mismo problema pero ocn una codificacion de la informacion diferente, con ewl fin de que el conjunton de datos sea el texto en crudo

```python
from tensorflow.keras.datasets import imdb

(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)
```

```python
# ГЌndices de las palabras
word_index = imdb.get_word_index()
```

```python
# Decodificando la reseГ±a utilizando los Г­ndices de las palabras de manera mГЎs eficientes
reversed_word_index = {v: k for k, v in word_index.items()}

" ".join([reversed_word_index.get(i - 3, "({})".format(i)) for i in train_data[0]])
```

```python
train_data[0]
```

### PreparaciГіn del conjunto de datos

```python
X_train_prep = []
X_test_prep = []
```

```python
# Preprocesando el conjunto de datos de entrenamiento
for review in train_data:
    X_train_prep.append(
        " ".join([reversed_word_index.get(i - 3, "") for i in review]))
```

```python
X_train_prep[0]
```

```python
# Preprocesando el conjunto de datos pruebas
for review in test_data:
    X_test_prep.append(
        " ".join([reversed_word_index.get(i - 3, "") for i in review]))
```

```python
X_test_prep[0]
```

```python
Y_train = train_labels
Y_test = test_labels
```

### DivisiГіn del conjunto de datos

Lo ideal es aplicar la divisiГіn del conjunto de datos antes de realizar ningГєn preprocesamiento para que no se infiera ningГєn tipo de tendencia como consecuencia de la transformaciГіn.

```python
from sklearn.model_selection import train_test_split

X_test_prep, X_val_prep, Y_test, Y_val = train_test_split(X_test_prep, Y_test, test_size=0.5)
```

```python
print("Longitud subconjunto entrenamiento: ", len(X_train_prep))
print("Longitud subconjunto validaciГіn: ", len(X_val_prep))
print("Longitud subconjunto pruebas: ", len(X_test_prep))
```

### VectorizaciГіn del conjunto de datos

```python
# esta clase lo que hace es hacer la vectorizacion del texto
# selecciona todas las palabras unicas que aparecen en el subconjunto de datos de entreenamiento (x_train)
# todas las palabras unicas son una caracteristica de entrada
# va rese;a por rese;a, en funcion de si esa palabra aparece en la rese;a o no
# en este caso es un vector de si las palabras aparecen o no en la rese;a, el otro era de frecuencia de las palabras
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train_prep)
```

```python
X_train = X_train.toarray()
```

```python
import pandas as pd

pd.DataFrame(X_train, columns=vectorizer.get_feature_names_out())
```

```python
X_test = vectorizer.transform(X_test_prep)
X_test = X_test.toarray()
```

```python
X_val = vectorizer.transform(X_val_prep)
X_val = X_val.toarray()
```

### Entrenamiento de la red neuronal artificial

```python
from tensorflow.keras import models
from tensorflow.keras import layers

model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(X_train.shape[1],)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
from tensorflow.keras import optimizers

model.compile(
    optimizer=optimizers.SGD(learning_rate=0.0005), 
    loss='binary_crossentropy',
    metrics=['accuracy', 'Precision']
)
```

```python
history = model.fit(
    X_train,
    Y_train,
    epochs=10,
    validation_data=(X_val, Y_val))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history)[['loss', 'val_loss']].plot(figsize=(10, 6))
plt.grid(True)
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history)[['accuracy', 'val_accuracy']].plot(figsize=(10, 6))
plt.grid(True)
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.show()
```

### EvaluaciГіn con el subconjunto de pruebas

```python
test_loss, test_acc, test_prec = model.evaluate(X_test, Y_test)
```

```python
print('test_acc:', test_acc)
print('test_prec:', test_prec)
```
